```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/tags/markers0to8.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
tag_size = 170
gap = 57
```

```python
tags = []
```

```python
tag_id = 0

for col in range(3):
    for row in range(3):
        x = col * tag_size + col * gap
        y = row * tag_size + row * gap
        tag = gray[x:x+tag_size, y:y+tag_size]
        tag = cv2.resize(tag, (9, 9))[2:-2, 2:-2] // 255
        tags.append(tag)
        tag_id += 1
```

```python
img = cv2.imread('../data/tags/markers9to17.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

for col in range(3):
    for row in range(3):
        x = col * tag_size + col * gap
        y = row * tag_size + row * gap
        tag = gray[x:x+tag_size, y:y+tag_size]
        tag = cv2.resize(tag, (9, 9))[2:-2, 2:-2] // 255
        tags.append(tag)
```

```python
tags = np.array(tags)
```

```python
for tag in tags:
    plt.imshow(tag)
    plt.show()
```
